## Initial Setup

```python
from bs4 import BeautifulSoup
import requests
```

## Connect to Steam webpage

```python
r = requests.get("https://store.steampowered.com/tags/en/Action/")
r.status_code
```

```python
html = r.content
```

```python
soup = BeautifulSoup(html, "lxml")
```
